```python
LeagueGoals= dataset['Goals']
League= dataset['League']

plt.title("Highest Goals scored by 1 player in each league")
plt.subplots_adjust(right= 6)
plt.bar(League,LeagueGoals)
```

## Variances of Data columns

```python
Matches_Played=dataset['Matches_Played']
Mins= dataset['Mins']
Shots= dataset['Shots']
OnTarget= dataset['OnTarget']




Goalvariance = np.var(LeagueGoals, ddof=1)
Matchvariance= np.var(Matches_Played, ddof=1)
Minsvariance= np.var(Mins, ddof=1)
Shotsvariance = np.var(Shots, ddof=1)
onTargetvariance = np.var(OnTarget, ddof=1)





print( "Variance of Matches played" ,Matchvariance)
print( "Variance of Goals" ,Goalvariance)
print( "Variance of Minutes" ,Minsvariance)
print( "Variance of Shots" ,Shotsvariance)
print( "Variance of On Target" ,onTargetvariance)
```

```python
plt.scatter(Mins, LeagueGoals)

plt.xlabel('Minutes Played')
plt.ylabel('Goals Scored')
plt.title('Scatter Plot: Goals Scored vs. Minutes Played')
plt.grid(True)

plt.show()

correlation= np.corrcoef(Mins, LeagueGoals)

# Print the results
print("Correlation Coefficient:", correlation)
```

##### Moderate positive correlation between the 2 data columns
